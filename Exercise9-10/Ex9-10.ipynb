    <h4>Question 04 (Project)</h4>
<p>
 Develop a sort_bookings function that organizes bookings based on their 'start 
date'. The function should offer options for both ascending and descending 
order sorting.
</p>

</div>

```python
## Questions 02, 03 and 04:

# Importing necessary liberaries
import datetime

# Dictionary to save booking information
bookings = {}
# Constant for total number of cars
total_cars = 10
# Variable to generate unique booking codes
code = 1

# ----------------Functions----------------

## ----------------Question 02----------------
## Function to display information of just one booking
## This function is calling in display, search_by_name, search_by_code, update_booking and sort_bookings functions
def print_booking(code, booking):
  print("\n" + 5*"*" + str(code) + 5 * "*")
  print(f"Name : {booking['name']}")
  print(f"Start date : {booking['start_date']}")
  print(f"End date : {booking['end_date']}")
  print(f"Days : {booking['days']}")


## ----------------Question 03----------------
## Function to change 'name', 'start_date' and 'end_date' of a booking
def update_booking(code_upd):
  if code_upd in bookings:
    name = input("Enter your new name: ")
    start_date = input("New start date (YYYY-MM-DD): ")
    end_date = input("New end date (YYYY-MM-DD): ")
    # Validate the provided dates
    is_valid, msg = validate_date(start_date)
    if not is_valid:
      print(msg)
      return
    is_valid, msg = validate_date(end_date)
    if not is_valid:
      print(msg)
      return
    # Convert string dates to datetime
    start_date = datetime.datetime.strptime(start_date, "%Y-%m-%d").date()
    end_date = datetime.datetime.strptime(end_date, "%Y-%m-%d").date()
    count_days = find_days(start_date, end_date)
    # Check if the end date comes before the start date
    if start_date > end_date:
      print("Error!")
      return
    # Update booking
    modified_booking = {
      "name" : name, 
      "start_date" : start_date,
      "end_date" : end_date,
      "days" : count_days
    }
    bookings[code_upd].update(modified_booking)
    print_booking(code_upd, bookings[code_upd])
    print("Done!")
  else:
    print("Not found!")


## ----------------Question 04----------------
## Function to organize bookings based on their 'start date'. 
## The function should offer options for both ascending and descending order sorting.
def sort_bookings(reverse=False): # default value: False ==> ascending order
  if len(bookings) != 0:
    # x[1] means the inner dictionary (value of the outer dictionary) that is a booking: 
    sorted_bookings = dict(sorted(bookings.items(), key= lambda x : x[1]['start_date'], reverse=reverse)) 
    ## display the sorted bookings
    for code, booking in sorted_bookings.items():
      print_booking(code, booking)
  else:
    print("Empty")


# Function to display help information
def help():
  help_text = """
    booking: Make a Booking
    display: Show all Bookings
    search: Search Booking
    cancel: Cancel Booking
    details: Show details of the project
    sort: Show sorted bookings based on their 'start date'
    update: Update 'name', 'start date' and 'end date' of a booking
    exit: exit\n"""
  print (help_text)

# Function to validate the date entered by the user
def validate_date(input_date):
  c = input_date.count("-")
  # Check format of date
  if c != 2 or len(input_date) != 10 or input_date[4] != "-" or input_date[7] != "-":
    return False, "Invalid Format (YYYY-MM-DD)"
  # Check if year, month, and day are all digits
  parts = input_date.split("-") # [2020, 12, 15]
  if not (parts[0].isdigit() and parts[1].isdigit() and parts[2].isdigit()):
    return False, "Should be numbers"
  # Validate the range of year, month, and day
  year, month, day = list(map(int, parts))
  if year < 1 or not (1 <= month <= 12) or not (1 <= day <= 31):
    return False, "Invalid date"
  return True, "Valid date"

## Recursive function to find number of reservation's days
def find_days(start_date, end_date):
  if start_date == end_date:
    return 0
  else:
    new_start_date = start_date + datetime.timedelta(days=1)
    return 1 + find_days(new_start_date, end_date)

# Function to handle booking of cars
def booking():
  global code # Declare code as global since we're modifying it
  # Check if all cars are already booked
  if len(bookings) >= total_cars:
    print("All cars are booked!")
    return
  name = input("Enter you name: ")
  start_date = input("Start date (YYYY-MM-DD): ")
  end_date = input("End date (YYYY-MM-DD): ")
  # Validate the provided dates
  is_valid, msg = validate_date(start_date)
  if not is_valid:
    print(msg)
    return
  is_valid, msg = validate_date(end_date)
  if not is_valid:
    print(msg)
    return
  # Convert string dates to datetime
  start_date = datetime.datetime.strptime(start_date, "%Y-%m-%d").date()
  end_date = datetime.datetime.strptime(end_date, "%Y-%m-%d").date()
  # Check if the end date comes before the start date
  if start_date > end_date:
    print("Error!")
    return
  count_days = find_days(start_date, end_date)
  # Create new booking
  new_booking = {
    "name" : name, 
    "start_date" : start_date,
    "end_date" : end_date,
    "days" : count_days
  }
  # save with new code
  bookings[code] = new_booking
  print_booking(code, bookings[code]) ## print the details of booking
  # Increment the booking code for the next booking
  code += 1
  print("Done!")

# Function to cancel a booking
def cancel(code_del):
  bookings.pop(code_del, "Not found!")

# Function to display all current bookings
def display():
  if len(bookings) != 0:
    for code, booking in bookings.items():
      print_booking(code, booking)
  else:
    print("Empty")

# Function to search for a booking using its code
def search_by_code(code_s):
  if code_s in bookings:
    booking = bookings[code_s]
    print_booking(code_s, booking)
  else:
    print("Not found!")

def search_by_name(name):
  found = False
  for code, booking in bookings.items():
    if booking["name"] == name:
      print_booking(code, booking)
      found = True
  if not found:
    print(f"{name}: Not found!")
  

def search():
  cmd = input("Search by 'name' or 'code': ")
  if cmd == "name":
    name = input("Name for search: ")
    search_by_name(name)
  if cmd == "code":
    code_s = int(input("Code for search: "))
    search_by_code(code_s)
  else:
    print("Not found!")

# Details to show a details of the cars
def details():
  num_booked_cars = len(bookings)
  num_available_cars = total_cars - num_booked_cars
  print(f"Total number of cars: {total_cars}")
  print(f"Number of booked: {num_booked_cars}")
  print(f"Number of availabel: {num_available_cars}")

# ----------------Main----------------
while True:
  command = input("\nEnter your option (help, booking, display, search, cancel, update, sort, details, exit): ")
  if command == "help":
    help()
  ## ----------------Question 03----------------
  elif command == "update":
    code_upd = int(input("Enter your code: "))
    update_booking(code_upd)
  ## ----------------Question 04----------------
  elif command == "sort":
    sorting_order = input("Enter '0' for ascending order or '1' for descending order: ")
    if sorting_order == '0' or sorting_order == '1' :
      sort_bookings(bool(int(sorting_order)))
    else:
      print("Invalid sorting order!")
  elif command == "booking":
    booking()
  elif command == "display":
    display()
  elif command == "search":
    search()
  elif command == "cancel":
    code_del = int(input("Enter your code: "))
    cancel(code_del)
  elif command == "details":
    details()
  elif command == "exit":
    break
  elif command == "":
    continue
  else:
    print("Command not found!")
```

<p>
Create a Python function to multiply all numbers between two user given integers, using recursion.
</p>

```python
def mul_int_range(num1, num2):
    if num1 == num2:
        return num1
    else:
        return num1 * mul_int_range(num1+1, num2)

num1 = int(input("Enter first integer number: "))
num2 = int(input("Enter second integer number: "))
num1, num2 = min(num1, num2), max(num1, num2)
mul = mul_int_range(num1, num2)
print(f"Product of all integers in range [{num1}, {num2}] is {mul}")
```
