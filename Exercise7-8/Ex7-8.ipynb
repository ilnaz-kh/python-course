    <h4>Question 01 (Function):</h4>
    <p>define two lists that one of them is strings and another is float, change values of 
both of them to int and create a list that is sum of them. use map function and 
lambda.
</p>
</div>

```python
## define summation inline function using lambda:
sum_two_num = lambda num1, num2 : num1 + num2
## define two lists:
str_list = ["12", "4", "06", "-3", "24"]
float_list = [2.4, 3., -4.5, 0.0]
## Change values of both list to int using map:
int_list1 = list(map(int, str_list))
int_list2 = list(map(int, float_list))
print(f"List1: {int_list1}")
print(f"List2: {int_list2}")
##sum of two lists:
sum_list = list(map(sum_two_num, int_list1, int_list2))
print(f"Sum of two lists: {sum_list}")
```

    <h4>Question 02 (Function)</h4>
<p>
 Write a Python program where users input two numbers and choose an 
operation (+, -, *, /, **). Use lambda functions for calculations. If division is 
chosen and the second number is zero, replace it with one, notify the user, and 
proceed. The program should handle invalid operations gracefully. This program 
should be repeated until user write “exit”.

</p>

</div>

```python
## Summation:
sum_ = lambda a, b : a+b
## Substract:
substract = lambda a, b : a-b
## Multiplication:
multiply = lambda a, b : a*b
## Division:
division = lambda a, b : a/b
## Power:
power = lambda a, b : a**b


while True:
    valid_operations = ["+", "-", "*", "/", "**"]
    op = input("\nEnter the operation (+, -, *, /, **, exit): ")
    if op == "exit":
        break
    elif op == "":
        continue
    elif op not in valid_operations:
        print(f"{op} : Invalid operation!")
    else: # valid operations
        ## get two operators
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if op == "+":
            print(f"{num1} + {num2} = {sum_(num1, num2)}")

        elif op == "-":
            print(f"{num1} - {num2} = {substract(num1, num2)}")

        elif op == "*":
            print(f"{num1} x {num2} = {multiply(num1, num2)}")

        elif op == "/":
            if num2 == 0:
                print("Cannot divide by zero")
                print(f"{num1} / 1 = {num1}")
            else:
                print(f"{num1} / {num2} = {division(num1, num2)}")

        else: # op == "**"
            print(f"{num1} ^ {num2} = {power(num1, num2)}")
            
```

    <h4>Question 03 (Function)</h4>

 Write question of exercise 6 by function.

```python

def add(task_name):
    if task_name not in task_names:  # task names are unique
        task_names.append(task_name)
        task_statuses.append(False)
        task_durations.append(None)
        print(f"Task '{task_name}' has been added!")
    else: # input task name exist in current tasks
        print(f"The name '{task_name}' already exists in the list!")

def display():
    if task_names: # task list is not empty
        print("#   Task  ,  Status  ,  Duration")
        for idx, task_name in enumerate(task_names):
            if task_statuses[idx] == True:
                task_status = "Completed"
                task_duration = task_durations[idx] # duration != None
            else:
                task_status = "Incomplete"
                task_duration = "-" # duration == None    
            print(f"({idx+1})  {task_name}  ,  {task_status}  ,  {task_duration}")
    else:
        print("There is no task in the list!")

def remove(task_name):
    if task_name in task_names:
        task_idx = task_names.index(task_name)
        task_names.pop(task_idx)
        task_statuses.pop(task_idx)
        task_durations.pop(task_idx)
        print(f"Task '{task_name}' has been removed!")
    else:
        print(f"There is no task named '{task_name}' to remove!")

def edit(old_task_name):
    if old_task_name in task_names:
        new_task_name = input("new task name: ")
        if new_task_name in task_names:
            print(f"The new name '{new_task_name}' already exists in the list!")
        else:
            task_idx = task_names.index(old_task_name)
            task_names[task_idx] = new_task_name
            print(f"The task name '{old_task_name}' has been changed to '{new_task_name}'!")
    else:
        print(f"There is no task named '{old_task_name}' to edit!")

def search(task_name):
    close_result = []
    similar_result = []
    for idx, name in enumerate(task_names):
        if name == task_name: # Exact Match
            close_result.append(idx)
        elif name.find(task_name) != -1 : # Similar
            similar_result.append(idx)

    ## Print Search results:
    if close_result:
        task_idx = close_result[0]
        task_name = task_names[task_idx]
        if task_statuses[task_idx] == True:
            task_status = "Completed"
            task_duration = task_durations[task_idx] # duration != None
        else:
            task_status = "Incomplete"
            task_duration = "-" # duration == None    
        print(f"Close Result:\nTask: {task_name} , Status: {task_status} , Duration: {task_duration}")

    if similar_result:
        print("Similar Result(s):")
        for task_idx in similar_result:
            task_name = task_names[task_idx]
            if task_statuses[task_idx] == True:
                task_status = "Completed"
                task_duration = task_durations[task_idx] # duration != None
            else:
                task_status = "Incomplete"
                task_duration = "-" # duration == None               
            print(f"Task: {task_name} , Status: {task_status} , Duration: {task_duration}")

    if not close_result and not similar_result : # both lists are empty
        print(f"No results found!")

def done(task_name):
    if task_name in task_names:
        task_idx = task_names.index(task_name)
        if task_statuses[task_idx]: # True
            print(f"Task '{task_name}' is already marked as done!")
        else:
            start = float(input("Enter start time: "))
            end = float(input("Enter end time: "))
            if start < 0 or end < 0 or start > 24 or end > 24:
                print("Start time or end time must be between 0 and 24!")
            elif end < start:
                print("The start time cannot be after the end time!")
            else:
                task_duration = end - start
                task_durations[task_idx] = task_duration
                task_statuses[task_idx] = True
                print(f"Task '{task_name}' is marked as done!")
    else:
        print(f"There is no task named '{task_name}'!")

def details():
    if task_names: # list is not empty
        num_tasks = len(task_names)
        num_completed = task_statuses.count(True)
        num_incomplete = num_tasks - num_completed
        hours_worked = 0 # initialize
        for duration in task_durations:
            if duration: # duration != None
                hours_worked += duration
        print(f"Number of Tasks: {num_tasks}\nNumber of Completed Tasks: {num_completed}\tNumber of Incomplete Tasks: {num_incomplete}")
        print(f"Total Hours Worked: {hours_worked}")
        ## display:
        print("#   Task  ,  Status  ,  Duration")
        for idx, task_name in enumerate(task_names):
            if task_statuses[idx] == True:
                task_status = "Completed"
                task_duration = task_durations[idx] # duration != None
            else:
                task_status = "Incomplete"
                task_duration = "-" # duration == None    
            print(f"({idx+1})  {task_name}  ,  {task_status}  ,  {task_duration}")
    else:
        print("There is no task in the list!")

def help():
    help = """
    Commands:
    - add: add a new task
    - display: display all tasks by details (task, status, duration)
    - remove: remove a task
    - edit: edit the name of an existing task
    - search: search for a task by name
    - done: mark a task as done
    - details: show details of all tasks
    - help: show the help script
    - exit: exit program
    """
    print(help)

################## Main

task_names = [] # string list
task_statuses = [] # boolean list
task_durations = [] # float list

for i in range(200):
    command = input("\nYour command (add, display, remove, edit, search, done, details, help, exit):  ")

    if command == "add":
        task_name = input("task name (to add): ")
        add(task_name)

    elif command == "display":
        display()

    elif command == "remove":
        if task_names:
            task_name = input("task name (to remove): ")
            remove(task_name)
        else:
            print("There is no task in the list!")

    elif command == "edit":
        if task_names:
            old_task_name = input("Task name (to edit): ")
            edit(old_task_name)
        else:
            print("There is no task in the list!")

    elif command == "search":
        if task_names:  # list is not empty
            task_name = input("task name (to search): ")
            search(task_name)
        else:
            print("There is no task in the list!")

    elif command == "done":
        if task_names:
            task_name = input("task name (to mark as done): ")
            done(task_name)
        else:
            print("There is no task in the list!")

    elif command == "details":
        details()

    elif command == "help":
        help()
        
    elif command == "":
        continue

    elif command == "exit":
        break

    else: # none of the valid commands
        print(f"'{command}' is not a value command!")
```

```python
# ## version 1: change the list to dictionary
# product = {
#     "name" : 'Apple iPhone',
#     "price" : 999.99,
#     "color" : 'Space Gray',
#     "released_date" : 'July 2021'
# }
## version 2:
product_keys = ["name", "price", "color", "released_date"]
product_values = ['Apple iPhone', 999.99, 'Space Gray', 'July 2021']
product = dict(list(zip(product_keys, product_values)))
print(f"The product dictionary: {product}")

## 1. the name of the product
name = product["name"]
print(f"1. The name of the product: {name}")

## 2. the price of the product
price = product["price"]
print(f"2.The price of the product: {price}")

## 3. the color of the product
color = product["color"]
print(f"3. The color of the product: {color}")

## 4. the released date of the product
released_date = product["released_date"]
print(f"4. The released date of the product: {released_date}")

## 5. change the color of the product to 'Silver'
product["color"] = 'Silver'
print(f"5. The new color of the product: {product['color']}")

## 6.Create a list with 4 products buy this structure
products = []
for i in range(4):
    name = input(f"Name of product {i+1}: ")
    price = float(input(f"Price of product {i+1}: "))
    color = input(f"Color of product {i+1}: ")
    released_date = input(f"Released date of product {i+1}: ")
    product = {
        "name" : name,
        "price" : price,
        "color" : color,
        "released_date" : released_date
    }
    products.append(product)
## print the list
print("\nProducts:")
for product in products:
    print(product)
```

    <h4>Question 02 (Dictionary)</h4>
<p>
create a dictionary named `student_grades` containing three students' names 
as keys and their grades as values. Perform the following:
</p>
<ul>
<li>
add two new students with their grades to `student_grades`, names are unique.
</li>
<li>
Delete any students from `student_grades` who have grades below 10. 
</li>
<li>
Get a name from the user If the user's name is in the dictionary of students, get 
a new grade and change the student's grade.</li>
<li>
Show the average scores of all students.
</li>
</ul>
</div>

```python
## create a dictionary named 'student_grades' containing three students' names as keys and their grades as values
student_grades = {
    "name1" : 15.0,
    "name2" : 10.5,
    "name3" : 8.25
}
# print(student_grades)

## add two new students with their grades to `student_grades`, names are unique.
for i in range(2):
    name = input("Student name (to add): ")
    if name in student_grades.keys():
        print("Name already exists!")
    else:
        grade = float(input("Grade: "))
        ## add this student to dictionary
        student_grades[name] = grade

## Delete any students from 'student_grades' who have grades below 10. 
for std_name,std_grade in list(student_grades.items()):
    if std_grade < 10 :
        student_grades.pop(std_name)
## print results
print(f"\nAfter removing grades less than 10: {student_grades}")

## Get a name from the user If the user's name is in the dictionary of students, get a new grade and change the student's grade.
name = input("\nEnter a student name to change his/her grade: ")
if name in student_grades.keys():
    grade = float(input(f"Enter a new grade for the student '{name}': "))
    student_grades[name] = grade
else:
    print(f"The name '{name}' is not a student name")
print(f"\nAfter changing the grade of a student: {student_grades}")

## Show the average scores of all students.
average = sum(student_grades.values())/len(student_grades) if student_grades else 0
print(f"\nAverage scores of all students is: {average}")
```
