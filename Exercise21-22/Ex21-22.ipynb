```python
## Ex 21
from abc import ABC, abstractmethod
class Person(ABC):
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def edit_profile(self, new_name=None, new_age=None):
        self.name = new_name or self.name
        self.age = new_age or self.age

    @abstractmethod
    def describe_role(self):
        pass

    def __str__(self) -> str:
        return f"Name: {self.name}, Age: {self.age}"
```

```python
## Ex 21
# from person import Person
from datetime import datetime

class Member(Person):
    id_count = 1111 

    def __init__(self, name, age):
        super().__init__(name, age)
        self.member_id = self.id_maker()
        self.books_borrowed = []

    @classmethod
    def id_maker(cls):
        new_id =  "M" + str(cls.id_count)
        cls.id_count += 1
        return new_id
    
    def borrow_book(self, book):
        if not book.check_availability():
            raise ValueError(f"Error: The book ({book.book_id}) is on loan!")
        else:
            self.books_borrowed.append(book)
            book.is_borrowed = True
            book.borrow_date = datetime.now()
            print(f"The book ({book.book_id}) has been borrowed by member ({self.member_id})")

    def return_book(self, book):
        if book.check_availability():
            raise ValueError(f"Error: The book ({book.book_id}) is NOT on loan!")
        elif not book in self.books_borrowed:
            raise ValueError(f"Error: The book {(book.book_id)} has not been checked out by you!")
        else:
            self.books_borrowed.remove(book)
            book.is_borrowed = False
            book.borrow_date = None
            print(f"The book {(book.book_id)} has  been returned by member ({self.member_id})")

    def describe_role(self):
        print(f"{self.name} with ID {self.member_id} is a Member!")

    def __str__(self):
        books_title = []
        for book in self.books_borrowed:
            books_title.append(book.title)
        return f"Member ID: {self.member_id}, {super().__str__()}, BorrowedBooks: ({' ,'.join(books_title)})"
```

```python
## Ex 21
# from person import Person

class Librarian(Person):
    id_count = 1111 

    def __init__(self, name, age): 
        super().__init__(name, age)
        self.librarian_id =  self.id_maker()

    @classmethod
    def id_maker(cls):
        new_id =  "L" + str(cls.id_count)
        cls.id_count += 1
        return new_id
    
    def describe_role(self):
        print(f"{self.name} with ID {self.librarian_id} is a Librarian!")

    def __str__(self):
        return f"Librarian ID: {self.librarian_id}, {super().__str__()}"

```

```python
## Ex 21
class Book:
    id_count = 1111 

    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.book_id = self.id_maker()
        self.__is_borrowed = False
        self.borrow_date = None

    @classmethod
    def id_maker(cls):
        new_id =  "B" + str(cls.id_count)
        cls.id_count += 1
        return new_id
    
    def check_availability(self):
        return not self.__is_borrowed
    
    @property
    def is_borrowed(self):
        return self.__is_borrowed
    
    @is_borrowed.setter
    def is_borrowed(self, new_status):
        self.__is_borrowed = new_status

    def __str__(self) -> str:
        if self.__is_borrowed:
            date_string = self.borrow_date.strftime("%m-%d-%Y, %H:%M:%S")
            return f"Book ID: {self.book_id}, Title: {self.title}, Author: {self.author}, Avaialability: No, BorrowingDate: {date_string}"
        else: 
            return f"Book ID: {self.book_id}, Title: {self.title}, Author: {self.author}, Avaialability: Yes"
 
```

```python
## Ex 21
from datetime import datetime

class ReservationError(Exception):
    def __init__(self, msg):
        self.msg = msg
        super().__init__(self.msg)

class Reservation:
    id_count = 1111 

    ## Method to create a new reservation object:
    def __init__(self, member, book):
        try: 
            self.valid_reservation(member, book)
            self.member = member
            self.book = book
            self.reservation_id = self.id_maker()
            self.date = datetime.now()
            self.status = True # is currently active
        except ValueError as e:
            raise e

    @staticmethod
    def valid_reservation(member, book):
        if book.check_availability():
            raise ReservationError("ReservationError: The requested book is NOT on loan!")
        if book in member.books_borrowed:
            raise ReservationError("ReservationError: The requested book has been checked out by you!")
        ## If a book is borrowed by another reader, you can reserve it:
        return True
        
    def cancle_reservation(self):
        self.status = False
        print(f"Reservation ({self}) successfully cancelled!")

    @classmethod
    def id_maker(cls):
        new_id =  "R" + str(cls.id_count)
        cls.id_count += 1
        return new_id
    
    def __str__(self):
        return f"Resevation ID: {self.reservation_id}, Date: {self.date.strftime('%m-%d-%Y, %H:%M:%S')}, Status: {self.status}"
    
```

```python
## Ex 22

import requests

class Music:
    def __init__(self, kind, artistName, collectionName, trackName, artworkUrl100, releaseDate, primaryGenreName ):
        self.kind = kind
        self.artistName = artistName
        self.collectionName = collectionName
        self.trackName = trackName
        self.artworkUrl100 = artworkUrl100
        self.releaseDate = releaseDate
        self.primaryGenreName = primaryGenreName

    def __str__(self) -> str:
        str1 = f"kind: {self.kind}, artistName: {self.artistName}, \ncollectionName: {self.collectionName}, "
        str2 = f"trackName: {self.trackName}, \nreleaseDate: {self.releaseDate}, primaryGenreName: {self.primaryGenreName}"
        return  str1 + str2 
                


class MusicFetcher():
    API_URL = "https://itunes.apple.com/search?entity=song&term="

    @staticmethod
    def fetch_music(term):
        response = requests.get(MusicFetcher.API_URL + term)
        if response.status_code == 200:
            music_items = response.json()
            music_items = music_items['results']
            musics = []
            for item in music_items:
                music_obj = Music(item.get('kind', ''), item.get('artistName', ''), item.get('collectionName', ''),
                                  item.get('trackName', ''), item.get('artworkUrl100', ''), 
                                  item.get('releaseDate', ''), item.get('primaryGenreName', ''))
                musics.append(music_obj)
            return musics
        else:
            print("Cannot connect to the API")
            return []
        
def main():
    term = "rainbow" # query term
    results = MusicFetcher.fetch_music(term)
    for music in results:
        print('-'*60)
        print(music)

if __name__ == "__main__":
    main()
```
