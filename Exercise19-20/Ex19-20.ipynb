```python
## Ex 19
class InsufficientFundsError(Exception):
    def __init__(self, bank_account, msg):
        self.bank_account = bank_account
        self.msg = f"Error: Account Number: {self.bank_account.account_number}, {self.bank_account.balance} : {msg}"
        super().__init__(self.msg)


class InvalidTransactionError(Exception):
    def __init__(self, amount, msg):
        self.amount = amount
        self.msg = f"Error: {self.amount} : {msg}"
        super().__init__(self.msg)


class BankAccount:
    ac_num = 1111
    def __init__(self, balance=0):
        if self.is_valid_balance(balance):
            self.__balance = balance
            self.__account_number = BankAccount.ac_num  # a unique account number
            print(f"Bank account opened successfully: Account Number: {self.__account_number}, Balance= {self.__balance}")
            BankAccount.ac_num += 1

    def deposit(self, amount):
        if amount <=0 :
            raise InvalidTransactionError(amount, "Cannot deposit a negative amount of money")
        self.__balance += amount
        print(f"Account Number: {self.__account_number}, Balance: {self.__balance}")

    def withdraw(self, amount):
        if amount > self.__balance:
            raise InsufficientFundsError(self, "Insufficient Balance!")
        self.__balance -= amount
        print(f"Account Number: {self.__account_number}, Balance: {self.__balance}")

    @property
    def balance(self):
        return self.__balance
    
    # @balance.setter
    # def balance(self, new_balance):
    #     if new_balance < 0 :
    #         raise ValueError("Balance must be a positive number")
    #     self.__balance = new_balance

    @property
    def account_number(self):
        return self.__account_number
    
    @staticmethod
    def is_valid_balance(balance):
        if balance < 0:
            raise InvalidTransactionError(balance, "Balance cannot be negative")
        return True


if __name__ == "__main__":
    try:
        ba1 = BankAccount()
        ba2 = BankAccount(120)
        ba3 = BankAccount(-3) # raise an error
    except Exception as e:
        print(e)

    try:
        ba1.withdraw(20) # raise an error
    except InsufficientFundsError as e:
        print(e)

    try:
        ba2.withdraw(20)
    except InsufficientFundsError as e:
        print(e)

    try:
        ba1.deposit(130)
    except InvalidTransactionError as e:
        print(e)

    try:
        ba2.deposit(-20) # raise an error
    except InvalidTransactionError as e:
        print(e)

```

```python
## Ex 20
from abc import ABC, abstractmethod
class Person(ABC):
    def __init__(self, name, age, id_):
        if not self.valid_identity_id(id_):
            raise ValueError("Error: Identity ID is a 5 digit number.")
        self.name = name
        self.age = age
        self.id_ = id_ # identity ID is getting from input

    @abstractmethod
    def introduce(self):
        pass
    
    def __str__(self):
        return f"Name: {self.name}, Identity ID: {self.id_}, Age: {self.age}"
    
    @staticmethod
    def valid_identity_id(id_):
        ## Suppose a valid identity ID has 5 digits (in string format)
        if id_.isdigit() and len(id_) == 5:
            return True
        return False    


class Librarian(Person):
    id_count = 1111 
    def __init__(self, name, age, id_): 
        super().__init__(name, age, id_)
        self.employee_id = "E" + str(Librarian.id_count) # employee_id is created by application
        Librarian.id_count += 1
    
    def introduce(self):
        return f"{self}, Employee ID: {self.employee_id}"  

   
class Member(Person):
    id_count = 1111 
    def __init__(self, name, age, id_):
        super().__init__(name, age, id_)
        self.member_id = "M" + str(Member.id_count) # member_id is created by application
        Member.id_count += 1

    def introduce(self):
        return f"{self}, Member ID: {self.member_id}"  


class Book:
    def __init__(self, title, author, isbn):
        if not self.valid_isbn(isbn):
            raise ValueError("Error: ISBN is a 4 digit number.")
        self.title = title
        self.author = author
        self.__isbn = isbn

    # @property
    # def isbn(self):
    #     return self.__isbn
    
    # @isbn.setter
    # def isbn(self, new_isbn):
    #     ## Suppose ISBN is a 4-digits number
    #     if not new_isbn.isdigit() or len(new_isbn) != 4:
    #         raise ValueError("Error: ISBN is a 4 digit number.")
    #     self.__isbn = new_isbn

    @staticmethod
    def valid_isbn(isbn_):
        ## Suppose ISBN is a 4-digits number String
        if isbn_.isdigit() and len(isbn_) == 4:
            return True
        return False  

    def __str__(self):
        return f"Title: {self.title}, Author: {self.author}"

    def introduce(self):
        return f"Title: {self.title}, Author: {self.author}, ISBN: {self.__isbn}"


class Library:
    def __init__(self, name):
        self.name = name
        self.entities = {
            "members" : {}, # (key, value) : (member's identity ID, Member object)
            "librarians" : {}, # (key, value) : (librarian's identity ID, Librarian object)
            "books" : {}, # (key, value) : (Book's ISBN, Book object)
        }

    def add_book(self, title, author, isbn):
        if isbn not in self.entities["books"]:
            try:
                new_book = Book(title, author, isbn) # raise a ValueError
                self.entities["books"][isbn] = new_book
                print(f"Book added successfully: Title: {title}, Author: {author}, ISBN: {isbn}")
            except ValueError as e:
                print(e)
        else:
            print(f"Error: ISBN '{isbn}' already exists!")

    def register_member(self, name, age, id_): # id_ is a 5-digits number
        if id_ not in self.entities["members"]:
            try:
                new_member = Member(name, age, id_) # raise a ValueError
                self.entities["members"][id_] = new_member
                print(f"Member registered successfully:  Name: {name}, Age: {age}, Identity ID: {id_}, Member ID: {new_member.member_id}")
            except ValueError as e:
                print(e)
        else:
            print(f"Error: Identity ID '{id_}' already exists!")

    def register_librarian(self, name, age, id_):
        if id_ not in self.entities["librarians"]:
            try:
                new_librarian = Librarian(name, age, id_) # raise a ValueError
                self.entities["librarians"][id_] = new_librarian
                print(f"Librarian registered successfully:  Name: {name}, Age: {age}, Identity ID: {id_}, Librarian ID: {new_librarian.employee_id}")
            except ValueError as e:
                print(e)
        else:
            print(f"Error: Identity ID '{id_}' already exists!")

    def display(self, type_="librarians"):
        if type_ not in self.entities:
            raise ValueError(f"\n'{type_}': Invalid Argument!")
        
        print("\n" + "-"*10 + type_ + "-"*10)
        if not self.entities[type_]:
            print("Not Found!")
        else:
            for entity in self.entities[type_].values():
                print(entity.introduce())

    def __str__(self) -> str:
        str1 = "\n" + "-"*10 + "Library's information" + "-"*10 + "\n"  
        str2 = f"#Members: {len(self.entities['members'])}, #Librarians: {len(self.entities['librarians'])}, #Books: {len(self.entities['books'])}\n"
        return  str1+str2


if __name__ == "__main__":
    lbr = Library("DehKhoda")
    
    try:
        # lbr.register_librarian("l1", 23, "12345")
        lbr.register_member("m1", 15, "13265")
        lbr.register_member("m2", 40, "54433")
        lbr.register_member("m3", 33, "n43r45") # raise a ValueError
    except ValueError as e:
        print(e)

    try:
        lbr.register_member("m4", 50, "54433") # raise a ValueError
    except ValueError as e:
        print(e)

    print(lbr)

    try:
        lbr.add_book("t1", "a1", "4532")
        lbr.add_book("t2", "a2", "4532") # raise a ValueError
    except ValueError as e:
        print(e)
    
    print(lbr)

    try:
        lbr.display("members") # "librarians" (default) , "members", "books"
        lbr.display() # "librarians" (default) , "members", "books"
        lbr.display("books") # "librarians" (default) , "members", "books"
        lbr.display("dummyArgument") # raise a ValueError
    except ValueError as e:
        print(e)
```
