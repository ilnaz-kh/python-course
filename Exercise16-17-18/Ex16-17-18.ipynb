```python
## Exercise 16 - Question 01
class Vehicle:
    ## class attribute:
    total_vehicles = 0

    def __init__(self, make, model, year) -> None:
        if self.is_valid_year(year): # accessing static method using 'self'
            self.make = make
            self.model = model
            self.year = year
            Vehicle.update_total_vehicles() # accessing class attribute using class name 'Vehicle'
            print(f"Instance '{make}', '{model}','{year}' is created successfully!")
        else:
            raise ValueError(f"Cannot create instance: Invalid year: {year}")
        
    ## instance method:
    def display_vehicle(self):
        print("--------------------------------------\nVehicle:")
        print(f"Make: {self.make}, Model: {self.model}, Year: {self.year}")
        print("--------------------------------------")

    @staticmethod
    def is_valid_year(year): # a static method
        if isinstance(year, int) and year > 1930 and year < 2024:
            return True
        return False

    @classmethod
    def update_total_vehicles(cls): # a class method which uses 'cls'
        cls.total_vehicles += 1 # accessing class attribute inside the class method using 'cls'

if __name__ == "__main__":
    try:
        v1 = Vehicle("m1", "model1", 2019)
        v2 = Vehicle("m1", "model2", 12) #invalid date
    except ValueError as e:
        print(e)
    v1.display_vehicle()
    print(f"Total vehicles: {Vehicle.total_vehicles}") # accessing class attribute outside of the class using class name 'Vehicle'
```

```python
## Exercise 17 - Question 01
class MyString(str):
    
    # def __init__(self,data=""):
    #     super().__init__(data)

    ## Returns a list of all indices where a given substring is found
    def find_all(self, sub_str, start=0):
        indices = []
        while True:
            found = super().find(sub_str, start)
            if found == -1:
                break
            indices.append(found)
            start = found + 1
        return indices
    
    ## Returns the number of vowels in the string
    def count_vowels(self):
        vowels = ["a", "e", "i", "o", "u"]
        tmp = super().lower() # type(tmp) : <class 'str'>
        count = 0
        for vowel in vowels:
            count += tmp.count(vowel)
        return count
    
    ## Checks if the string is a palindrome
    def is_palindrome(self):
        return self == self[::-1]
    
    ## Change the upper method so that it only converts consonants to uppercase, leaving vowels in lowercase
    def upper(self):
        vowels = ["a", "e", "i", "o", "u"]
        string = "" # an empty 'str'
        for char in self: # type(char) : <class 'str'>
            if char in vowels:
                string += char
            else:
                string += char.upper()
        return(MyString(string)) # Casting

if __name__ == "__main__":

    sub = "P"
    s2 = MyString("Python Python Python")
    s3 = MyString("PPPPPPPPPPPPPPPPPPP")
    print(f"All occurances of '{sub}' in '{s2}': {s2.find_all(sub, 0)}")
    print(f"All occurances of '{sub}' in '{s3}': {s3.find_all(sub, 0)}")

    s4 = MyString("ABCabcIOU")
    print(f"Count of vowels in '{s4}': {s4.count_vowels()}")

    s5 = MyString("abcdef")
    print(f"Is {s5} palindrome? {s5.is_palindrome()}")
    s6 = MyString("abccba")
    print(f"Is {s6} palindrome? {s6.is_palindrome()}")

    s7 = MyString("hI python") # return a 'MyString' object
    print(f"Upper case of {s7}: {s7.upper()}")
    print(f"Type({s7}): {type(s7)}")
```

Develop a Python-based university management system to demonstrate your 
understanding of inheritance, polymorphism, and abstraction. Create classes 
for Student, Teacher, and University with specific attributes and methods.

```python
## Exercise 18 - Question 01
## understanding of inheritance, polymorphism, and abstraction

## 'Person' ia an abstract class with one attribute 'name' and one abstract method 'show_info()'
## 'Student' is a subclass of the 'Person' with additional attributes 'id_s' and 'courses' (inheritance)
## 'Teacher' is a subclass of the 'Person' with additional attributes 'id_t' and 'taught_courses' (inheritance)
## method 'show_info' is overwrited in both 'Student' and 'Teacher' classes. This indicates class polymorphism.

from abc import ABC, abstractmethod

## abstract class:
class Person(ABC): # abstraction
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def show_info(self): # polymorphism
        pass


## Student class inherits from Person class:
class Student(Person): # inheritance

    def __init__(self, name, id_s, courses=None):
        super().__init__(name)
        self.id_s = id_s
        if not courses:
            self.courses = []
        else:
            self.courses = courses

    def enroll(self, course):
        if course not in self.courses:
            self.courses.append(course)
            print(f"Course '{course}' successfully added to Student '{self.id_s}'!")
        else:
            print(f"Course {course} already exists!")

    def show_info(self): # polymorphism
        print(f"ID: {self.id_s}, Name: {self.name}, courses: {', '.join(self.courses)}")


## Teacher class inherits from Person class:
class Teacher(Person): # inheritance

    def __init__(self, name, id_t, taught_courses=None):
        super().__init__(name)
        self.id_t = id_t
        if not taught_courses:
            self.taught_courses = []
        else:
            self.taught_courses = taught_courses

    def teacher_course_assignment(self, course):
        if course not in self.taught_courses:
            self.taught_courses.append(course)
            print(f"Course '{course}' successfully added to teacher '{self.id_t}'!")
        else:
            print(f"Course {course} already exists!")

    def show_info(self): # polymorphism
        print(f"ID: {self.id_t}, Name: {self.name}, courses: {', '.join(self.taught_courses)}")


class University:
    
    def __init__(self):
        self.students = {} # an empty dictionary
        self.teachers = {} # an empty dictionary

    def add_student(self, name, id_s, courses=None):
        if id_s not in self.students:
            new_std = Student(name, id_s, courses)
            self.students[id_s] = new_std
            print(f"Student '{name}' Successfully added!")
        else:
            print(f"Student ID '{id_s}' already exists!")

    def add_teacher(self, name, id_t, courses=None):
        if id_t not in self.teachers:
            new_teacher = Teacher(name, id_t, courses)
            self.teachers[id_t] = new_teacher
            print(f"Teacher '{name}' Successfully added!")
        else:
            print(f"Teacher ID '{id_t}' already exists!")

    def students_details(self):
        if self.students:
            print("-----------------------------\nStudent details:")
            for std in self.students.values():
                std.show_info()
            print("-----------------------------")
        else:
            print("There is no student")

    def teachers_details(self):
        if self.teachers:
            print("-----------------------------\nTeacher details:")
            for teacher in self.teachers.values():
                teacher.show_info()
            print("-----------------------------")
        else:
            print("There is no teacher")

    def enrollment(self, id_s, course):
        if id_s in self.students:
           self.students[id_s].enroll(course)
        else:
            print(f"No student ID {id_s} found!")
            
    def course_assigning(self, id_t, course):
        if id_t in self.teachers:
            self.teachers[id_t].teacher_course_assignment(course)
        else:
            print(f"No teacher ID {id_t} found!")


def main():
    uni = University()
    
    uni.add_student("s1", 123)
    uni.add_student("s2", 134, ["History, Physics"])
    uni.students_details()
    uni.enrollment(134, "Math")
    uni.enrollment(134, "Chemistry")
    uni.enrollment(123, "Biology")
    uni.students_details()

    uni.add_teacher("t1", 234)
    uni.add_teacher("t2", 235)
    uni.teachers_details()
    uni.course_assigning(234, "Math")
    uni.teachers_details()

if __name__ == "__main__":
    main()
```
