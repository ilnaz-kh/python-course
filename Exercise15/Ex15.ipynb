```python
## Exercise 15 - Question 01
"""
class attribute vs. instance attribute
----------------------------------------------------------------------------
Define:
Class attributes are defined directly inside the class body. 
Instance attributes are defined inside the '__init__' method or other instance methods using 'self'.

Class attributes have one value across the class and all objects of the class. They are like a global variable in a functional programming.
Instance attributes can have different values for each instance. (Each instance has its own value for an instance attribute)

Inside the class: 
    in the instance method, a class attribute is accessed by <ClassName>.<classAttribute>
    in the class method, a class attribute is accessed by cls.<classAttribute>
    an instance variable is accessible just by 'self' (inside the instance methods)
Outside the class:
    a class attribute can be accessed by both instance and class name: <InstanceName>.<classAttribute> or <ClassName>.<classAttribute>
    instance attribute can be accessed just by instance name, that return the attribute's value for that specific object.

    
class method vs. instance method
----------------------------------------------------------------------------
instance methods have a 'self' argument, that refers to the calling object.
inside the class they can be called using self.<InstanceMethod>. outside of the class they are accessed by object name and '.'

class methods are created using @classmethod decorator. They have a 'cls' argument that refer to the class itself.
Outside of the class, they can be called using both class name and object name as following:
<ClassName>.<ClassMethod>() or <InstanceName>.<ClassMethod>()

"""

class Student:

    def __init__(self, std_id, std_name):
        ## instance attributes:
        self.std_id = std_id
        self.std_name = std_name

    def __str__(self) -> str:
        return f"ID: {self.std_id}, Name: {self.std_name}"


class Students:
    ## define class attribute:
    num_of_stds = 0

    def __init__(self):
        ## define instance attributes:
        self.stds = {}

    ## define instance method:
    def add_std(self, s_id, s_name):
        ## Access instance attribute 'stds' inside the instance method:
        if s_id not in self.stds: 
            std = Student(s_id, s_name)
            self.stds[s_id] = std
            ## Access class attribute 'num_of_stds' inside the instance method:
            Students.num_of_stds += 1
            print(f"{std} added successfully!")

    ## define class method:
    @classmethod
    def get_num_of_stds(cls):
        ##  Access class attribute inside the class method:
        return cls.num_of_stds   
         
## Outside of the class:
if __name__ == "__main__":

    students = Students()
    ## Access instance attribute 'stds':
    print(students.stds)
    ## Calling instance method 'add_std':
    students.add_std("123", "a")
    students.add_std("124", "b")
 
    ## Access class attribute using instance:
    print(students.num_of_stds)
    ## Access class attribute using class name:
    print(Students.num_of_stds)

    ## Calling class method using instance:
    print(students.get_num_of_stds())
    ## Calling class method using class name:
    print(Students.get_num_of_stds())
```

```python
## Exercise 15 - Question 02
from datetime import *
class Task:
    def __init__(self, task_id, name) -> None:
        self.task_id = task_id
        self.name = name
        self.state = False
        self.date_of_finishing = None

    def __str__(self) -> str:
        if self.state == True:
            return f"ID: {self.task_id}, Name: {self.name}, State: Complete, Date of Fininshing: {self.date_of_finishing}"
        else:
            return f"ID: {self.task_id}, Name: {self.name}, State: Incomplete, Date of Fininshing: -"

    def done(self):
        self.state = True
        self.date_of_finishing = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    def undone(self):
        self.state = False
        self.date_of_finishing = None

## class test:
if __name__ == "__main__":
    t1 = Task("1", "a")
    print(t1)
    t1.done()
    print(t1)
```

```python
## Exercise 15 - Question 02
# from task import *
class ToDoList:
    def __init__(self) -> None:
        self.tasks = {} # (key, value):  (task.task_id, task) 

    def add_task(self, task_id, name):
        if task_id not in self.tasks:
            new_task = Task(task_id, name)
            self.tasks[task_id] = new_task
            print("Task is added!")
        else:
            print("Task already exists!")

    def remove_task(self, task_id):
        if task_id in self.tasks:
            del self.tasks[task_id]
            print("Task is removed!")
        else:
            print("Not found!")

    ## Editing name and state of the task
    def edit_task(self, task_id):
        if task_id in self.tasks:
            new_name = input("New name (press Enter to skip): ")
            self.tasks[task_id].name = new_name or self.tasks[task_id].name
            new_state = input("Enter 'C' to mark the task as completed one or 'I' to mark it incomplete (press anything else to skip): ")
            if new_state.lower() == "c":
                self.tasks[task_id].done()
            elif new_state.lower() == "i":
                self.tasks[task_id].undone()
            print("The edited task:")
            print(self.tasks[task_id])
        else:
            print("Not found!")

    ## partial matching search:
    def search_by_name(self, search_name):
        found = False
        for task in self.tasks.values():
            if search_name in task.name: # partial matching
                print(task)
                found = True
        if not found:
            print("Not found!")

    def search_by_id(self, search_id):
        if search_id in self.tasks:
            print(self.tasks[search_id]) # print a 'Task' object
        else:
            print(f"ID: {search_id} not found!")

    def search_by_state(self, search_state):
        found = False
        for task in self.tasks.values():
            if task.state == search_state:
                print(task)
                found = True
        if not found:
            print("Not found!")

    def display_all(self):
        if len(self.tasks): # there is at least one task
            for task in self.tasks.values():
                print(task) # call `__str__` method of class `Task`
        else:
            print("There is no task!")

## class test:
if __name__ == "__main__":
    manager = ToDoList()
    manager.display_all()
    manager.add_task("1", "a")
    manager.add_task("2", "b")
    manager.display_all()
    manager.remove_task("3")
    manager.remove_task("1")
    manager.display_all()
    manager.edit_task("2")
```

```python
## Exercise 15 - Question 02
# from todolist import *

def main():
    manager = ToDoList()
    while True:
        cmd = input("add, remove, edit, search, display, exit: ")
        if cmd =="add":
            task_id = input("ID: ")
            name = input("name: ")
            manager.add_task(task_id, name)
        elif cmd== "remove":
            task_id = input("Task ID to remove: ")
            manager.remove_task(task_id)
        elif cmd == "edit":
            task_id = input("Task ID to edit: ")
            manager.edit_task(task_id)
        elif cmd == "search":
            search_type = input("Enter 'N' for Search by Name, 'I' for search by ID or 'S' for search by State: ")
            search_type = search_type.lower()
            if search_type == "n":
                search_name = input("Name to search: ")
                manager.search_by_name(search_name)
            elif search_type == "i":
                search_id = input("ID to search: ")
                manager.search_by_id(search_id)
            elif search_type == "s":
                search_type = input("Enter 'C' for completed tasks or 'I' for incomplete ones: ")
                if search_type.lower() == "c":
                    manager.search_by_state(True)
                elif search_type.lower() == "i":
                    manager.search_by_state(False)
                else:
                    print("Invalid Input")
            else:
                print("Invalid input")
        elif cmd == "display":
            manager.display_all()
        elif cmd == "exit":
            break
        elif cmd == "":
            continue
        else:
            print(f"{cmd}: Not found!")

if __name__ == "__main__":
    main()
```

```python
## Exercise 15 - Question 03

class BookClub:
    ## class attribute:
    total_members = 0
    ## initialization:
    def __init__(self, club_name) -> None:
        ## instance attributes:
        self.club_name = club_name
        self.members = {} # an empty dictionary (key: value) = (member_name, member_favorite_book)

    def add_member(self, name, book):
        if name not in self.members:
            self.members[name] = book
            BookClub.total_members += 1 # update total count
            print(f"Name: {name} successfully added!")
        else:
            print(f"Name {name} already exists!")

    def remove_member(self, name):
        if name in self.members:
            del self.members[name]
            BookClub.total_members -= 1
            print(f"Name: {name} successfully removed!")
        else:
            print(f"Name {name}: Not found!")
    
    def display_members(self):
        if len(self.members):
            print("Member(s):")
            for name, book in self.members.items():
                print(f"Name: {name},   Favorite Book: {book}")
        else:
            print("Not found!")
        
    @classmethod
    def get_total_members(cls):
        return cls.total_members
    
## testing class:
if __name__ == "__main__":
    bookclub = BookClub(club_name="Sun") # initialize a BookClub instance

    ## display all members
    print(f"Number of Members: {BookClub.get_total_members()}")
    bookclub.display_members()

    ## add members
    bookclub.add_member(name= "a", book="Python")
    bookclub.add_member(name= "b", book="C++")
    bookclub.add_member(name= "a", book="Java")

    ## display all members
    print(f"Number of Members: {BookClub.get_total_members()}")
    bookclub.display_members()

    ## remove
    bookclub.remove_member(name= "c")
    bookclub.remove_member(name="a")

    ## display all members
    print(f"Number of Members: {BookClub.get_total_members()}")
    bookclub.display_members()
```
